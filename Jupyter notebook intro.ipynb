# Введение в Jupyter notebook

Ноутбук (notebook) - документ специального формата [JSON](https://ru.wikipedia.org/wiki/JSON) с расширением .ipynb, который генерируется приложением Jupyter Notebok App и одновременно содержит программный код и расширенной текст, включающий заголовки, параграфы, уравнения, графики, ссылки. Ноутбуки представляют удобно читаемый текст с аналитическим описанием и результатами (таблицами, графиками, формулами) и исполняемый программный код, который может быть запущен в браузере. [Официальное описание доступно по ссылке.](https://jupyter-notebook.readthedocs.io/en/stable/)

Jupyter Notebook App - это клиент-серверное приложение, которое позволяет редактировать и запускать ноутбуки в браузере. Jupyter Notebook приложение может быть запущено на локально на рабоем компьютере или на удаленном сервере и доступно в браузере через интернет. Кроме возможность отображения/редактирования/выполнения ноутбуков приложение включает dashboard - панель управления, показывающую файлы на локальном компьютере, позволяющую открыть ноутбуки или завершить их процессы (kernels). Приложение Jupyter Notebook устанавливается вместе с дистибутивом python для научных вычислений Anaconda.

Процесс ноутбука (kernel) - это "вычислительный движок", который исполняет код ноутбука. Ipython kernel отвечает за код python, но существуют движки и для дугих языков программирования. При открытии ноутбука автоматически запускается связанный с ним kernel, при выполнении команд ноутбука движок проводит вычисления и возвращает результаты, потребляя ресурсы CPU и RAM. При завершении процесса память очищается.

Запусить приложение Jupyter Notebook можно кликнув на иконку в навигаторе Anaconda, напечатав команду jupyter notebook в терминале, либо запустив файл Jupyter Notebook в папке Scripts. После старта приложение предоставляет доступ к файлам в стартовом каталоге.

При закрытии браузера (или вкладки) прилоежение Jupyter Notebook не завершает свою работу. Для полного завершения работы требуется закрыть терминал, где запущен процесс Jupyter Notebook. Приложение Jupyter Notebook App - это сервер, который отображается в браузере по адресу http://localhost:8888, завершение работы браузера не завершает рабоут сервера. Можно (но не рекомендуется) запускать несколько серверов Jupyter Notebook. Для завершения процесса (kernel) можно выполнить команду File -> Close and Halt в меню ноутбука либо в панели управления (Dashboard) на вкладке Running завершить работу процесса.

Запустить ноутбук можно в панели управления Jupyter Notebook. Для пошагового выполнения ноутбука на каждой ячейке (cell) нужно нажать shift + enter. Для выполнения всего ноутбука за один шаг требуетсяс выбрать меню Cell -> Run All. Перезапустить вычислительный движок можно командой меню Kernel -> Restart, при этом вся память освобождается, программа выполняется "с чистого листа".

## Текстовые ячейки

Для добавления текста в ноутбук используются ячейки в режиме разметки (Markdown cells). Изменть тип ячейки можно в меню Cell панели инструментов или сочетанием клавиш выбрав ячеку и нажав "m". Язык разметки Markdown - популярный инструмент для структурирования текста, явлется надстройкой языка HTML, все теги HTML отображаются в Markdown формате. [Руководство по оформлению Markdown файлов](https://gist.github.com/Jekins/2bf2d0638163f1294637)

Для *курсивого* написания текст нужно окружить "\*", для **утолщения** шрифта текст требуется окружить "\**".

Markdown поддерживает упорядоченные (нумерованные) и неупорядоченные (ненумерованные) списки. Для формирования неупорядоченный списков используются такие маркеры, как звездочки, плюсы и дефисы.
- элемент 1
- элемент 2
- элемент 3

Для формирования упорядоченных списков в качестве маркеров используются числа с точкой. Важной особенностью в данном случае является то, что сами номера, с помощью которых формируется список, не важны, так как они не оказывают влияния на выходной HTML код. Как бы ни нумеровал пользователь список, на выходе он в любом случае будет иметь упорядоченный список, начинающийся с единицы (1, 2, 3…). 
1. элемент 1
4. элемент 2
7. элемент 3

Вложенные списки организуются отступами (табуляциями):
- первый уровень
    + второй уровень
        + третий уровень
        - третий уровень
    - второй уровень
- первый уровень

Горизонтальную разделительную линию между блоками текста можно создать при помощи символов "\*\*\*" или "---".

> Пример цитаты

**%env** - позволяет управлять переменными среды для вашего блокнота без перезапуска Jupyter-сервера. Некоторые библиотеки (такие, как theano) используют переменные среды, чтобы контролировать поведение, и %env — самый удобный способ.


В Notebook можно вызвать любую shell-команду. Это особенно удобно для управления виртуальной средой, например,
> !pip install numpy

````python
!pip install numpy
````

Иногда вывод не нужен, и в этом случае можно или использовать команду pass с новой строки, или поставить точку запятой в конце строки:

````python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy
plt.hist(numpy.linspace(0, 1, 1000)**1.5);
````

**%run** может выполнить код на языке Python из файлов с расширением .py — это поведение хорошо задокументировано. Но эта команда может выполнять и другие блокноты из Jupyter! Иногда это очень полезно.
Обратите внимание, что %run — это не то же, что импорт python-модуля.

````python
%run ode.ipynb
````

%store — ленивая передача данных между блокнотами. 

````python
data = 'this is the string I want to pass to different notebook'
%store data
del data # deleted variable
````

````python
# В другом ноутбуке
%store -r data
print(data)
````
