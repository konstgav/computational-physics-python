# Распределения молекул идеального газа

### 1. Распределение Максвелла по проекции скорости

$$\varphi (v_x) = \sqrt{\frac{m}{2\pi kT}}\exp^{-\frac{mv_x^2}{2kT}} $$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
m = 2*1.66e-27 # кг
k = 1.38e-23 # Дж/К
T = 300. # К
v_max = 10000. # м/c
N_steps = 1000
vx = np.linspace(-v_max, v_max, N_steps)
phi_cold = np.sqrt(m/(2*np.pi*k*T))*np.exp(-m*vx**2/(2*k*T))
T = 1000. # К
phi_hot = np.sqrt(m/(2*np.pi*k*T))*np.exp(-m*vx**2/(2*k*T))
plt.title("Распределение Максвелла по проекции скорости")
plt.xlabel(r"$v_x$")
plt.ylabel(r"$\varphi(v_x)$")
plt.plot(vx, phi_cold, 'b', label=r"$T = 300$ K")
plt.plot(vx, phi_hot, 'r', label=r"$T = 1000$ K")
plt.legend()
plt.grid()
plt.show()
```

### 2. Распределение Максвелла по модулю скорости

$$f(v) = 4\pi v^2 \left( \frac{m}{2\pi kT} \right)^{\frac{3}{2}} \exp^{-\frac{mv^2}{2kT}} $$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
m = 2*1.66e-27 # кг
k = 1.38e-23 # Дж/К
T = 300. # К
v_max = 10000. # м/c
N_steps = 1000
v = np.linspace(0, v_max, N_steps)
f_cold = 4.*np.pi*v**2*(m/(2*np.pi*k*T))**(1.5)*np.exp(-m*v**2/(2*k*T))
T = 1000. # К
f_hot = 4.*np.pi*v**2*(m/(2*np.pi*k*T))**(1.5)*np.exp(-m*v**2/(2*k*T))
plt.title("Распределение Максвелла по проекции скорости")
plt.xlabel(r"$v$")
plt.ylabel(r"$f(v)$")
plt.plot(v, f_cold, 'b', label=r"$T = 300$ K")
plt.plot(v, f_hot, 'r', label=r"$T = 1000$ K")
plt.legend()
plt.grid()
plt.show()
```
