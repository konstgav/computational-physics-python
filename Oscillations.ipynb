## Колебания

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
```

$$x=A\cos(\omega t +\varphi_0)$$
$$v=-\omega A \sin(\omega t +\varphi_0)$$
$$a=-\omega^2 A \cos(\omega t +\varphi_0)$$

```python
t = np.linspace(0, 10, 1000)
A = 1.
omega = 10.
phi0 = 0.
x = A*np.cos(omega*t+phi0)
v = -omega*A*np.sin(omega*t+phi0)
a = -omega**2*A*np.cos(omega*t+phi0)
```

```python
fig, axs = plt.subplots(3)
axs[0].plot(t,x,'b') #синий
axs[0].set_title(r'$x$')
axs[1].plot(t,v,'g') #зеленый
axs[1].set_title(r'$v$')
axs[2].plot(t,a,'r') #красный
axs[2].set_title(r'$a$')
fig.tight_layout()
```

## Сложение колебаний
$$x_1 = A\cos (\omega t + \varphi_1), x_2 = A\cos(\omega t + \varphi_2),$$
$$x_1 + x_2 = 2A \cos (\omega t + \frac{\varphi_1 + \varphi_2}{2}) \cos \frac{\varphi_1 - \varphi_2}{2}.$$

```python
phi1 = 0
phi2 = np.pi
x1 = A*np.cos(omega*t+phi1)
x2 = A*np.cos(omega*t+phi2)
```

```python
fig, axs = plt.subplots(3)
axs[0].plot(t,x1,'b') #синий
axs[0].set_title(r'$x_1$')
axs[1].plot(t,x2,'g') #зеленый
axs[1].set_title(r'$x_2$')
axs[2].plot(t, x1 + x2,'r') #красный
axs[2].set_title(r'$x_1 + x_2$')
fig.tight_layout()
```

## Биения
Близкие частоты: $\Delta \omega \ll \omega$.  $$x_1 = A \cos \omega t, x_2 = A \cos(\omega + \Delta \omega) t,$$
$$x_1 + x_2 = 2A \cos (\omega + \frac{\Delta \omega}{2}) t \cos \frac{\Delta \omega}{2}t.$$

```python
omega = 10.
delta = 1.
x1 = A*np.cos(omega*t)
x2 = A*np.cos((omega+delta)*t)
plt.plot(t, x1 + x2)
plt.show()
```

## Фигуры Лиссажу
$$x = A \cos k\omega t, y = A \cos (l\omega t + \varphi).$$

```python
omega = 10.
phi = np.pi/4.
k=3
l=2
x = A*np.cos(k*omega*t)
y = A*np.cos(l*omega*t + phi)
plt.plot(x,y)
plt.show()
```

## Сложение колебаний с несоизмеримыми частотами

```python
omega1 = 2.
omega2 = 5.
t = np.linspace(0, 40, 10000)
x1 = A*np.cos(omega1*t)
x2 = A*np.cos(omega2*t)
plt.plot(t, x1 + x2)
plt.show()
```

## Фурье-анализ колебаний

```python
from scipy.fft import rfft, rfftfreq
SAMPLE_RATE  = 100
DURATION = 20
t = np.linspace(0, DURATION, SAMPLE_RATE*DURATION)
freq1 = 3.
freq2 = 5.
x1 = A*np.cos(2*np.pi*freq1*t)
x2 = 2*A*np.cos(2*np.pi*freq2*t)
plt.title("Сигнал")
plt.plot(t, x1+x2)
plt.show()

yf = rfft(x1+x2)
xf = rfftfreq(SAMPLE_RATE*DURATION, 1./SAMPLE_RATE)
plt.grid()
plt.title("Фурье-образ сигнала")
plt.plot(xf,np.abs(yf))
plt.show()
```
