# Задача Бюффона о бросании иглы

Бюффон впервые предложил вероятностный способ определения числа $\pi$. Суть метода заключается в случайном бросании на пол, составленный из равномерно расположенных половиц (досок), иглы. Определить вероятность того, что игла приземлится на линию, разделяющую половицы, можно по частоте пересечения иглой горизонтальных линий. Можно показать, что эта вероятность напрямую связана с пи. В особом случае, когда половые доски в два раза шире длины иголок, вероятность пересечения иглой границы досок составляет всего $1/\pi$. Таким образом, моделирование эксперимента Бюффона может использоваться (хотя и неэффективно) в качестве [метода Монте-Карло](https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D1%82%D0%BE%D0%B4_%D0%9C%D0%BE%D0%BD%D1%82%D0%B5-%D0%9A%D0%B0%D1%80%D0%BB%D0%BE) для аппроксимации числа $\pi$.

Напишите программу на python, которая рассчитывает число $\pi$ используя алгоритм, описанный Бюффоном. В качестве параметров моделирования можно задать следующие глобальные константы:

 - needlesNum = 10000 — число бросаний иглы в течение симуляции
 - needleLength = 1 — длина иглы
 - floorboardWidth = 2 — расстояние между половицами (досками)
 - xMax = yMax = 10 — размер области, где производится бросание иглы
